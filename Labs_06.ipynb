```python
#11. Write a Python program to check if a set is a subset of another set and print corresponding answer
s1 = {10,20,30,40}
s2 = {30,40,50,60,70}
s3 = {30,40}

print("If s3 is subset of s2")
print(s3.issubset(s2))

print("If s1 is superset of s3")
print(s1.issuperset(s3))
```

```python
#12. Write a Python program to create set difference
s1 = {10,20,30,40}
s2 = {30,40,50,60,70}
s3 = s2 - s1
print(s3)
```

```python
#13. Write a Python program to create a symmetric difference
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6, 7, 8}
s3 = s2 ^ s1
print(s3)
```
