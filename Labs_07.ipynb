```python
#3. Calculate factorial of a number using function
a = int(input("Enter a number: "))

def fact(a):
    total = 1
    for i in range(1,a+1):
        total *= i
    return total

print(f"The factorial of {a} is", fact(a))
```

```python
#4. Create sequence of Fibonacci using function
n = int(input("Enter a number: "))

def fibonacci(n):
    a = 0
    b = 1
    if n == 1:
        print(f"The fibionacci series upto {n} terms is", a, end="")
    elif n==2:
        print(f"The fibionacci series upto {n} terms is", a,b, end = "")
    else:
        print(f"The fibionacci series upto {n} terms is", a,b, end = " ")
        for i in range(n-2):
            c = a+b
            print(c, end=" ")
            a = b
            b = c
        
        
fibonacci(n)
```

```python
#5. Write a program to swapping of two numbers using functions
a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
print(f"The numbers before swapping are: {a}, {b}")

def swap(a,b):
    a,b = b,a
    return a,b
print("The numbers after swapping")
print("The numbers are swaped",swap(a,b))
```

```python
#6. Write a function to find the HCF of some given numbers
a = int(input("Enter a number: "))
b = int(input("Enter a number: "))

def HCF(a,b):
    if a>b:
        small = b
    else:
        small = a
    for i in range(1,small+1):
        if ((a%i==0)and(b%i==0)):
            hcf = i
    return hcf
print(f"The HCF of {a} and {b} is", HCF(a,b))
```

```python
#11. Write a program that demonstrates keyword arguments
def all_total(*args):
    total = 0
    for i in args:
        total += i
    print(type(args))
    return (total)
print("The sum is: ",all_total(1,2,3,4,5))
```
